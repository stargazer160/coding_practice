출처 백준 1316번

그룹 단어 체커


그룹 단어란 단어에 존재하는 모든 문자에 대해서, 각 문자가 연속해서 나타나는 경우만을 말한다. 예를 들면, ccazzzzbb는 c, a, z, b가 모두 연속해서 나타나고, kin도 k, i, n이 연속해서 나타나기 때문에 그룹 단어이지만, aabbbccb는 b가 떨어져서 나타나기 때문에 그룹 단어가 아니다.

단어 N개를 입력으로 받아 그룹 단어의 개수를 출력하는 프로그램을 작성하시오.

첫째 줄에 단어의 개수 N이 들어온다. N은 100보다 작거나 같은 자연수이다. 둘째 줄부터 N개의 줄에 단어가 들어온다. 단어는 알파벳 소문자로만 되어있고 중복되지 않으며, 길이는 최대 100이다.

```python
n = int(input()) # 첫째 줄에 단어의 개수 N이 들어온다. N은 100보다 작거나 같은 자연수이다.

num_group_word = 0  # 그룹단어의 개수를 출력하는 변수 지정 
for _ in range(n): # for 변수 in range(n==단어의개수==반복할횟수):
                   #     반복할 코드
    word = input() # 둘째 줄부터 N개의 줄에 단어가 들어온다. 단어는 알파벳 소문자로만 되어있고 중복되지 않으며, 길이는 최대 100이다.
    error = 0 # 각 문자가 떨어져서 나타나는 횟수에 대한 변수지정 
    for index in range(0, len(word)-1) : # 인덱스 범위 지정 : 0부터 단어의 문자열개수-1까지(단어의 문자의 개수만큼)
        if word[index] != word[index+1] :  # 만약 연속한 두 문자가 다르다면 ,
            new_word = word[index+1: ]  # 현재 문자의 다음 문자열을 새로운 단어 지정한후 
            if new_word.count(word[index]) > 0:  # 남아있는 문자열과 동일한 알파벳이 있는지를 확인
                error += 1  # 남아있는 문자열과 동일한 알파벳이 있다면 error에 1씩 증가.
    if error == 0:  # error가 0이면 그룹단어
        num_group_word += 1  # error가 0인 경우에 그룹단어의 개수 + 1 
print(num_group_word) # 그룹단어의 개수 출력 
```

출처 백준 1193번 : 분수찾기 

https://www.acmicpc.net/problem/1193

 * 분수의 배열

*   1번라인  : 1/1 (순방향, 개수 1)
*   2번라인  : 1/2, 2/1 (분자 : 오름차순, 분모: 내림차순 ,역방향, 개수 2)
*   3번라인  : 3/1, 2/2 ,1/3 (분자: 내림차순, 분모: 오름차순, 순방향, 개수 3 )
*   4번라인  : 1/4, 2/3, 3/2, 4/1 (분자 : 오름차순, 분모: 내림차순, 역방향, 개수 4)
*   5번라인  : 5/1, 4/2, 3/3, 2/4, 1/5 (분자: 내림차순, 분모: 오름차순,순방향,개수5)




```python
x=int(input())  # 정수 x가 주어짐 
line = 1  # 입력받은 정수(변수 x)가 몇번째 라인인지 표시 
while True: # x번째 분수는 어느 그룹에 있는가?
    if x < = sum(range(1,line+1)): # sum(range(1, line+1))은 1 + 2 + 3.... +line 임 
        break
    line + = 1 # 라인에 1을 더함
line_seq = x - sum(range(1,line))  # 그 그룹의 몇번째에 있는가? sum(range(1, line))은 1 + 2 ....+ line-1 
if line % 2 == 0: # 사선라인이 짝수 라인일때
    a = line_seq #분모
    b = line + 1 - line_seq #분자
else: # 사선라인이 홀수 라인 일때 
    a = line + 1 - line_seq #분모  
    b = line_seq #분자
print("{}/{}".format(a,b)) # " / " 형식으로 출력 


```
