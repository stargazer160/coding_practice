```python
word = input()
alpabets= [ 'c=','c-','dz=','d-','lj','nj','s=','z=']
for i in alpabets : 
  word = word.replace(i, '*')  # replace('바꿀문자열', '새문자열')은 문자열 안의 문자열을 다른 문자열로 바꿉니다(문자열 자체는 변경하지 않으며 바뀐 결과를 반환합니다).
print(len(word)) # len함수는 문자열의 개수를 셉니다. 
```
