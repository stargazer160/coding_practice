1.   손익 분기점 




화면 캡처 2022-05-31 092431.png



*   손익분기점은 총수입 > 총비용인 시점 
*   총비용 = 고정비용 + 가변비용 * 판매량
*   총수입 = 가격 * 판매량 

    a + bx = cx 

*    a + (b-c)x =0
*    (b-c)x = -a

*    x = -a //(b-c)





  
   



*  x(손익분기점) = a//c-b

```python
a, b, c = map(int,input().split()) # 정수 a,b,c 가 주어짐 
if b >= c : # 고정비용이 없다고 가정해도 가변비용*판매량 >= 가격*판매량이면 손익분기점 존재하지 않음 
  print(-1) # 손익분기점 존재하지 않으면 -1 출력 
else: # 손익분기점이 존재한다면 
  print(a//(c-b)+1) # 이익이 발생하는 판매량 = 손익분기점 판매량 + 1 
  
```

2. 벌집

화면 캡처 2022-05-31 112044.png

```python
n = int(input()) # 주어진 숫자 n
cnt = 1 # 숫자n이 지나가는 방의 개수
num = 1 # 방의 번호 
while  n > num :
   num += 6 * cnt # 방의 번호는 6의 배수만큼 증가함 
   cnt += 1 # 지나가는 방의 개수는 1씩 증가 
print(cnt) 
```


*   1 = 1
*   2 = 1 + (6 * 0)
*   8 = 2 + (6 * 1)
*   20 = 8 + (6 * 2)
*   38 = 20 + (6 * 3)
