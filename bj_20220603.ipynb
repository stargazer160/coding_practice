2. 설탕공장 

*   출처 : https://www.acmicpc.net/problem/2839




화면 캡처 2022-06-05 121747.png

화면 캡처 2022-06-06 144322.png

```python
n = int(input())
bags = 0 
while True:
  if (n % 5) == 0 :
      bags = bags + (n//5)
      print(bags)
      break
  n = n - 3 
  bags += 1       
  if n < 0 : 
     print(-1)
     break
```
