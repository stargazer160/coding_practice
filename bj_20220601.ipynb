1.   부녀회장이 될테야

*  https://www.acmicpc.net/problem/2775

화면 캡처 2022-06-05 121651.png

화면 캡처 2022-06-05 133013.png

```python
t = int(input()) # 테스트케이스의 수 

for _ in range(t): # 테스트 케이스 만큼 반복 
  k = int(input()) # 층수
  n = int(input()) # 호수 
  k_0 = [i for i in range(1,n+1)] # 0층 리스트(1,2..n)
  for k in range(k): # 층수만큼 반복 
     for i in range(1, n): # 리스트 인덱싱을 위한 인덱스로 사용 
      k_0[i] += k_0[i-1] # 층별 각 호실의 사람수를 변경 
  print(k_0[-1]) # 가장 마지막수 출력 
```
