달팽이는 올라가고 싶다

*   https://www.acmicpc.net/problem/2869


화면 캡처 2022-05-31 130503.png

```python
a, b, v = map(int,input().split()) # map 함수 = map(function, iterable)
# map(적용시킬 함수, 적용할 값들(반복가능한 자료형(리스트,튜플등)))
length = v - b # 정상에 올라가면 b만큼 다시 내려가야 하므로 나무막대의 높이는 v-b 임 
if (v - b) % (a - b) == 0 : # (나무막대의 높이/ 하루에 올라가는 높이)의 나머지가 0이면
    print(( v - b) // (a - b)) # 날짜를 출력한다.
else :
    print((( v - b )// (a - b)) +1) # 나머지가 0이 아니라면, 하루가 더 필요한 것으로 보고 +1 해준다.
```

화면 캡처 2022-05-31 145222.png

```python
t = int(input())
for i in range(t): 
  h, w, n = map(int,input().split())
  floor = n % h  # 층수
  number = (n // h)+ 1 # 호수 
  if n % h == 0 : # n이 h의 배수일때 
    floor = h # 나머지가 없으므로 층수는 h
    number = n//h      
  print(f'{floor*100+number}')
```
